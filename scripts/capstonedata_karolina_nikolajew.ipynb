```python
matches_list = []     #finding desc values with HOMICIDE in it

for item in san['Incident Description']:
    if re.findall(r'\b(homicide|murder|manslaughter|death)\b', item, flags =re.I):
        matches_list.append(item)

matches_list = list(dict.fromkeys(matches_list)) # removing duplicates while keeping order        

for match in matches_list:
    print(match)
```
